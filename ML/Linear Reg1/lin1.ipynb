```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model as lnr 
```

```python
df=pd.read_excel('data1.xlsx',index_col=0)
df.head()
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area,df.Price,color='r',marker='+')
plt.show()
```

```python
reg=lnr.LinearRegression()
reg.fit(df[['Area']],df[['Price']])
```

```python
reg.predict([[7500]])
```

```python
dff=pd.read_excel('test.xlsx') 
dff.head()
```

```python
dff['Price']=reg.predict(dff[['Area']]).astype(int)
```

```python
dff.head()
```

```python
dff.to_csv('test_prediction.csv')
```
