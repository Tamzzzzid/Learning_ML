```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model as lnr 
```

```python
df=pd.read_excel('home_prices.xlsx',index_col=0)
df.head()
```

```python
df.info()
```

```python
dff=df[['area','bedrooms']]
dff=dff.dropna()
```

```python
regt=lnr.LinearRegression()
regt.fit(dff[['area']],dff.bedrooms)
```

```python
df['bedrooms'] = df['bedrooms'].fillna(pd.Series(regt.predict(df[['area']])))
```

```python
df.info()
```

```python
#Multi dimantion linear reg 
```

```python
reg=lnr.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.predict([[3333,4,7]]).astype(int)
```

```python
df['predicted_prices'] = reg.predict(df[['area', 'bedrooms', 'age']]).astype(int)
```

```python
df.head()
```

```python
df['error']=(((df['price']-df['predicted_prices'])/df['price'])*100).astype(int).astype(str)+"%" 
```

```python
df.head()
```

```python
df.to_csv('home_prices_predicted.csv', index=False)
```
