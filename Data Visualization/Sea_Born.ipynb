```python
import seaborn as sns
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%config InlineBackend.figure_format = 'retina'
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
data.info()
```

```python
data['smoker'].value_counts()
```

```python
fig=sns.histplot(data=data,x='total_bill',kde=True)
fig.set_title('Bill graph');
```

```python
bins=np.arange(0,52,3)
fig2=sns.histplot(data=data,x='total_bill',kde=True,bins=bins)
fig2.set_title('Bill graph')
fig2.grid(True)
fig2.set(xticks=bins,yticks=bins)
fig2;
```

```python
sns.jointplot(data=data,x='total_bill',y='tip')
```

```python
sns.jointplot(data=data,x='total_bill',y='tip',kind='hex')
```

```python
sns.pairplot(data=data)
```

```python
sns.pairplot(data=data,hue='sex')
```

```python
sns.countplot(data=data ,x='smoker')
```

```python
sns.countplot(data=data ,x='smoker',hue='sex')
```

```python
# in bar plot we need to compare categorical data with numerical data , so one of the data must need to be categorical
# and other one should be numerical
```

```python
sns.barplot(data=data,x='day',y='total_bill')
```

```python
fig,axe=plt.subplots(1,2,figsize=(8,4)) 
sns.barplot(data=data,x='sex',y='total_bill',estimator=np.mean,ax=axe[0])
axe[0].set_title('Average bill')
sns.barplot(data=data,x='sex',y='total_bill',estimator=np.sum,ax=axe[1])
axe[1].set_title('total bill')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=data,x='day',y='total_bill')
```

```python
sns.violinplot(data=data,x='day',y='total_bill',hue='sex')
```

```python
sns.stripplot(data=data,x='day',y='total_bill',hue='sex')
```

```python
sns.swarmplot(data=data,x='day',y='total_bill')
```

```python
sns.violinplot(data=data,x='day',y='total_bill')
sns.swarmplot(data=data,x='day',y='total_bill')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
sns.stripplot(data=data, x='day', y='total_bill', ax=axes[0])
axes[0].set_title('Strip Plot')
sns.boxplot(data=data, x='day', y='total_bill', ax=axes[1])
axes[1].set_title('Box Plot') 
sns.swarmplot(data=data, x='day', y='total_bill', ax=axes[2])
axes[2].set_title('Swarm Plot')
sns.violinplot(data=data,x='day',y='total_bill',ax=axes[3])
sns.swarmplot(data=data, x='day', y='total_bill', ax=axes[3])
axes[3].set_title('Swarm+Violin Plot')
plt.tight_layout() 
plt.show()
```

```python
flights=pd.DataFrame(data=sns.load_dataset('flights'))
flights.head()
```

```python
piv=flights.pivot_table(
    index='month', 
    columns='year', 
    values='passengers'
)
piv
```

```python
sns.heatmap(data=piv, annot=True, fmt='.0f',lw=0.1)
plt.tight_layout()
```
