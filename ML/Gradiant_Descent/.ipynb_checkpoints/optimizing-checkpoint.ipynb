```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model as lnr 
```

```python
df=pd.read_csv('test_prediction.csv',index_col=0)
df.head()
```

```python
reg=lnr.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
m=reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
def grad_desc(x,y,learning_r,itr_max,m_curr=0,c_curr=0) : 
    n=len(x)
    itr=0
    cost=0
    while True: 
        itr=itr+1
        y_pred=x*m_curr+c_curr
        dm=-(2/n)*sum(x*(y-y_pred))
        dc=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-(learning_r*dm)
        c_curr=c_curr-(learning_r*dc)
        cost=1/n *(sum([val**2 for val in (y-y_pred)]))
        if( (itr>itr_max) | (cost<1)) :
            break;
        
    print(f"ITERATION = {itr} m= {m_curr} c={c_curr} Cost = {cost}")
    reg.coef_=m_curr
    reg.intercept_=c_curr
```

```python
x=np.array(df['Area'])
y=np.array(df['Price'])
```

```python
grad_desc(x,y,0.01,100000,m,c)
```
