```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model as lnr 
```

```python
#test_prediction.csv
df=pd.read_csv('test_prediction.csv',index_col=0)
df.head()
```

```python
reg=lnr.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[5100]]).astype(int)
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area,df.Price,color='r',marker='+')
plt.plot(df.Area,reg.predict(df[['Area']]))
```
