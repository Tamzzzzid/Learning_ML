```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x=np.random.randint(1,40,12)
y=x*2+6 
```

```python
x=np.sort(x)
y=np.sort(y)
```

```python
plt.plot(x,y)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('yaxis')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b')

plt.subplot(2, 2, 2)
plt.plot(x, y * 0.5, 'r')

plt.subplot(2,2,3)
plt.plot(x*0.5,y**0.5,'m')

plt.subplot(2,2,4)
plt.plot(x,y**2,'y')

plt.show()
```

```python
plt.plot(x, y**1.5, 'b',label='line1')
plt.plot(x, y**1.4, 'r')
plt.plot(x**1.1,y,'m')
plt.plot(x,y**1.6,'y')
plt.legend()
plt.show()
```

```python
#oop approach 
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.85,0.85])
ax1.plot(x,y,'r+-')
ax1.set_title("Main Curve")
ax1.set_xlabel('idc')
ax1.set_ylabel('yuc')

ax2=fig.add_axes([0.175,0.4,0.37,0.37])
ax2.set_title('XD')
ax2.plot(x**0.5,y**1.5,'b+-')
plt.show()
```

```python
x = np.random.randint(1, 40, 12)
x = np.sort(x)  
y = x * 2 + 6
fig,ax=plt.subplots(ncols=2,nrows=2,figsize=(7,7))
ax[0,0].plot(x,y,'r+-')
ax[0,1].plot(x,y**2,'y+-')
ax[1,0].plot(x**2,y,'b+-')
ax[1,1].plot(x,-y**2,'g+-')
plt.show()
```

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,0.9,0.9])
ax.plot(x,y,'r+-',label='line1')
plt.show()
```

```python
plt.style.use('_mpl-gallery')
np.random.seed(1)
x=4+np.random.normal(0,1.5,200)
fig=plt.figure(figsize=(7,7))
ax=fig.add_axes([0,0,0.9,0.9])
ax.hist(x,bins=8,lw=0.5,edgecolor="white")
ax.set(xlim=(0,8),ylim=(0,56),xticks= np.arange(0,8),yticks=np.arange(0,56,9))
plt.show()
```

```python
plt.style.use('_mpl-gallery')
np.random.seed(1)
x=4+np.random.normal(0,2,250)
y=4+np.random.normal(0,2,len(x))
fig=plt.figure(figsize=(7,7))
ax=fig.add_axes([0,0,0.9,0.9])
ax.scatter(x,y,c=x>y, cmap='viridis')
ax.set(xlim=(0,8),ylim=(0,8),xticks= np.arange(0,8),yticks=np.arange(0,8))
plt.show()

```
