```python
import numpy as np 
import pandas as pd 
```

```python
data=np.arange(6,11)
data
```

```python
pd_series1=pd.Series(data=data)
pd_series1
print('2nd index : ')
pd_series1[1]
```

```python
indx=np.array([chr(i) for i in range(ord('a'),ord('f'))])
indx
```

```python
pd_series1.index=indx
pd_series1
```

```python
print('2nd index : ')
pd_series1['b']
```

```python
pd_series2=pd.Series(data=data*2 ,index=indx) 
pd_series2
```

```python
dic={
    'a' : 'Tamzid', 
    'b' : 'Ronaldo', 
    'c' : 'Totty'
}
pd_series3=pd.Series(data=dic) 
pd_series3
print('2nd index : ') 
pd_series3['b']
```
