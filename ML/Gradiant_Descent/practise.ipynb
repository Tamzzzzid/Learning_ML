```python
import pandas as pd 
import numpy as np 
import sklearn.linear_model as lnr
```

```python
df=pd.read_excel('test.xlsx') 
df.head()
```

```python
x=np.array(df['X_bmi'])
y=np.array(df['y_target'])
len(x)
```

```python
def grad_desc(x,y,learning_r,itr_max) : 
    m_curr=0
    c_curr=0
    n=len(x)
    itr=0
    cost=0
    while True: 
        itr=itr+1
        y_pred=x*m_curr+c_curr
        dm=-(2/n)*sum(x*(y-y_pred))
        dc=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-(learning_r*dm)
        c_curr=c_curr-(learning_r*dc)
        cost=1/n *(sum([val**2 for val in (y-y_pred)]))
        if( (itr>itr_max) | (cost<1)) :
            break;
        
    print(f"ITERATION = {itr} m= {m_curr} c={c_curr} Cost = {cost}")
```

```python
grad_desc(x,y,0.1,10000)
```

```python
reg=lnr.LinearRegression()
reg.fit(x,y)
```
