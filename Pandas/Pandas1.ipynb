```python
level1=['groupA']*3 + ['groupB']*3
level2=['r1','r2','r3','r4','r5','r6'] 
col=['c1','c2','c3']
level1
```

```python
index=list(zip(level1,level2))
index
```

```python
indx=pd.MultiIndex.from_tuples(tuples=index)
```

```python
df=pd.DataFrame(
    data=np.random.rand(6,3), 
    columns=col, 
    index=indx
)
df
```

```python
df.index.names=['group','row'] 
df
```

```python
df.xs('groupA')
```

```python
df.xs('r2', level='row')
```

```python
df.xs('groupA')
df.xs('r2', level='row')
```
