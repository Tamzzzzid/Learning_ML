```python
left = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2', 'K3'],
    'A': np.random.randint(1, 10, size=4),
    'B': np.random.randint(1, 10, size=4)
})
right=pd.DataFrame(
    data={
        'key1' : ['K0', 'K2', 'K4', 'K5'], 
        'A' : np.random.randint(10,20,4), 
        'B' : np.random.randint(20,40,4)
    }
    
)
```

```python
left 
```

```python
right
```

```python
pd.merge(left,right,on='key1',how='inner')
```

```python
pd.merge(left,right,on='key1',how='outer')
```

```python
pd.merge(left,right,on='key1',how='left')
```

```python
left = pd.DataFrame(
    index=[1,2,3,4],
    data={
    
    'A': np.random.randint(1, 10, size=4),
    'B': np.random.randint(1, 10, size=4)
})
right=pd.DataFrame(
    index=[3,4,5,6],
    data={
        'C' : np.random.randint(10,20,4), 
        'D' : np.random.randint(20,40,4)
    }
    
)
```
